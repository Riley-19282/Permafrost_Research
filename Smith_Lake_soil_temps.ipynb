#yearly avgs + arrangement

```python
#find yearly average for each depth at both sites
#SL1
SL1surf_avg = SL1surf[['Time', 'SL1:0.01m']] #subframe
SL1savg = SL1surf_avg.set_index('Time')
SL1savg.index = pd.to_datetime(SL1savg.index, format='%m/%d/%Y %H:%M')
SL1_surf_yravg = SL1savg.resample('y').mean()

SL1core_avg = SL1core[['Time','SL1:0.04m','SL1:0.14m','SL1:0.29m','SL1:0.54m','SL1:0.59m','SL1:0.64m','SL1:0.69m','SL1:0.74m','SL1:0.79m']]
SL1cavg = SL1core_avg.set_index('Time')
SL1cavg.index = pd.to_datetime(SL1cavg.index, format='%m/%d/%Y %H:%M')
SL1_core_yravg = SL1cavg.resample('y').mean()
# print(SL1_core_yravg)

SL1deep_avg = SL1deep[['Time','SL1:1.5m','SL1:2.0m','SL1:3.0m']]
SL1davg = SL1deep_avg.set_index('Time')
SL1davg.index = pd.to_datetime(SL1davg.index, format='%m/%d/%Y %H:%M')
SL1_deep_yravg = SL1davg.resample('y').mean()
# print(SL1_deep_yravg)

#SL4
SL4core_avg = SL4core[['Time','SL4:0.02m','SL4:0.10m','SL4:0.17m','SL4:0.25m','SL4:0.33m','SL4:0.40m','SL4:0.48m','SL4:0.55m','SL4:0.71m','SL4:0.86m','SL4:1.07m']]
SL4cavg = SL4core_avg.set_index('Time')
SL4cavg.index = pd.to_datetime(SL4cavg.index, format='%m/%d/%Y %H:%M')
SL4_core_yravg = SL4cavg.resample('y').mean()
# print(SL4_core_yravg)

SL4deep_avg = SL4deep[['Time','SL4:1.0m','SL4:2.0m','SL4:2.8m', 'SL4:0.01m']]
SL4davg = SL4deep_avg.set_index('Time')
SL4davg.index = pd.to_datetime(SL4davg.index, format='%m/%d/%Y %H:%M')
SL4_deep_yravg = SL4davg.resample('y').mean()
# print(SL4_deep_yravg)
```
